```python
# Your code goes here

import pandas as pd
import numpy as np

url1 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
url2 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
url3 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"

df1 = pd.read_csv(url1)
df2 = pd.read_csv(url2)
df3 = pd.read_csv(url3)

# Combine datasets
combined_df = pd.concat([df1, df2, df3], ignore_index=True)
combined_df
```

```python
# Merge the 'ST' and 'STATE' columns into a single 'state' column
combined_df['state'] = combined_df['ST'].combine_first(combined_df['State'])

# Merge the 'GENDER' and 'gender' columns into a single 'gender' column
combined_df['gender'] = combined_df['GENDER'].combine_first(combined_df['Gender'])

# Drop the original 'ST', 'STATE', 'GENDER' columns
combined_df = combined_df.drop(columns=['ST', 'State', 'GENDER','Gender'])

# Show the cleaned DataFrame
print(combined_df)
```

```python
# Check for duplicates
print(combined_df.duplicated().sum())
```

```python
# Check for missing values
print(combined_df.isnull().sum())
```

```python
# Function to clean column names and drop NaN values from a DataFrame
def clean_column(df):
    """
    Cleans the column names of a DataFrame and drops rows with NaN values.
    - Converts all column names to lower case.
    - Replaces white spaces in column names with underscores.
    - Replaces 'st' with 'state' in column names.
    - Drops all rows with NaN values.

    Args:
        df (pd.DataFrame): The DataFrame to clean column names and remove NaNs for.

    Returns:
        pd.DataFrame: The cleaned DataFrame with no NaN rows.
    """
    # Apply transformations to column names
    df.columns = (
        df.columns
        .str.lower()                    # Convert to lower case
        .str.replace(' ', '_')          # Replace spaces with underscores
    )
    df.columns = ['state' if col == 'st' else col for col in df.columns]
    # Drop rows with NaN values
   

    return df
```

```python
cleaned_df = clean_column(combined_df)
```

```python
cleaned_df
```

```python
# Standardize gender column
def standardize_gender(df):
    df_cleaned = df.copy()
    
    # Ensure the gender column is of string type before applying string operations
    df_cleaned['gender'] = df_cleaned['gender'].astype(str)
    
    df_cleaned['gender'] = (
        df_cleaned['gender']
        .str.upper()  # Convert to uppercase
        .replace({'FEMALE': 'F', 'FEMAL': 'F', 'MALE': 'M'})  # Map values
    )
    
    # Handle unknown or invalid gender values
    df_cleaned['gender'] = df_cleaned['gender'].apply(lambda x: x if x in ['M', 'F'] else 'Unknown')
    
    return df_cleaned
```

```python
state_mapping = {
    'Oregon': 'OR',
    'California': 'CA',
    'Cali': 'CA',
    'Arizona': 'AZ',
    'Nevada': 'NV',
    'Washington': 'WA'
}

# Apply mapping to the 'state' column
cleaned_df['state'] = cleaned_df['state'].replace(state_mapping)
```

```python
# Education column standardization
cleaned_df['education'] = cleaned_df['education'].str.lower()  # Convert to lowercase for case-insensitive replacement
cleaned_df['education'] = cleaned_df['education'].replace('bachelors', 'bachelor') 
```

```python
combined_df['customer_lifetime_value'] = (
    combined_df['customer_lifetime_value']
    .astype(str)  # Convert to string to allow .str operations
    .str.replace('%', '', regex=False)  # Remove '%' character
    .astype(float)  # Convert back to float
)

cleaned_df['customer_lifetime_value'] = pd.to_numeric(cleaned_df['customer_lifetime_value'])
```

```python
def clean_open_complaints(df):
  """
  Cleans the 'Number of Open Complaints' column.

  Args:
    df: The DataFrame to clean.

  Returns:
    The DataFrame with the cleaned 'Number of Open Complaints' column.
  """
  df['number_of_open_complaints'] = (
      df['number_of_open_complaints']
      .str.split('/')
      .str[1]  # Extract the middle part
      .astype(float)  # Convert to float
      .fillna(0)  # Replace missing values with 0
  )
  return df
    
cleaned_df = clean_open_complaints(cleaned_df)
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
# Create a pivot table to summarize total revenue by sales channel
pivot_table = df4.pivot_table(
    values='total_claim_amount',  # This is the total revenue column
    index='sales_channel',        # This is the sales channel column
    aggfunc='sum',                # Summing the total revenue
    margins=True                  # Adds a total row
)

# Round the total revenue to 2 decimal points
pivot_table = pivot_table.round(2)

# Show the resulting pivot table
print(pivot_table)
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
# Create a pivot table to summarize the average customer lifetime value by gender and education
pivot_table = df4.pivot_table(
    values='customer_lifetime_value',  # This is the column for customer lifetime value
    index=['gender', 'education'],     # Group by gender and education
    aggfunc='mean'                     # Calculate the average
)

# Round the average customer lifetime value to 2 decimal points
pivot_table = pivot_table.round(2)

# Show the resulting pivot table
print(pivot_table)
```
